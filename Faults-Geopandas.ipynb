```python
gdf = gpd.read_file("CHAF_2020_allfiles/CHAF_Pangaea_v1/CHAF_Pangea_v1.shp")

geo_columns = [
        "F_id",
        "F_system",        
        "F_name",
        "strike",
        "length_km",
        "lat",
        "long",
        "type",
        "activity",
        "geometry"
]

for col in gdf.columns:
    if (col not in geo_columns):
        gdf.pop(col)

gdf["radian_strike"] = gdf["strike"].astype(float).apply(math.radians)
gdf["sin_strike"] = gdf["radian_strike"].apply(math.sin)
```
